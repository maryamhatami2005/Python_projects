- for loop on n characters (after conversion to string)
- power to 2
- sum 
- check sum == 1
- check if n is not in seen numbers

```python
n = 19
def is_happy(n):
    seen_numbers = set()
    while (n != 1) and (n not in seen_numbers):
        seen_numbers.add(n)
        n = sum([int(i) ** 2 for i in str(n)])
    
    return n == 1
    
```

```python
is_happy(19)
```

```python
def is_happy(x):
    seen_numbers = set()

    while x != 1 and x not in seen_numbers:
        seen_numbers.add(x)
        y = 0

        for digit in str(x):
            y += int(digit) ** 2
        x = y

    return x == 1
```
