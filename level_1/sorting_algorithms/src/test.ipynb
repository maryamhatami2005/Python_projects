```python
%%timeit

def bubble_sort(list):
    n = len(list)
    for i in range(n):
        for j in range(n-1-i):
            if list[j] > list[j+1]:
                list[j], list[j +1 ] = list[j + 1], list[j]
    return list

list_nums = [1,7,3,5,9,6]
bubble_sort(list_nums)
```

```python
%%timeit

def insertion_sort(list):
    n = len(list)
    for i in range(1,n):
        key = list[i]
        j = i - 1
        while j >= 0 and list[j] > key:
            list[j + 1] = list[j]
            j -= 1
        list[j + 1] = key
    return list
list_nums = [1,7,3,5,9,6]
insertion_sort(list_nums)
```

```python
def selection_sort_pythonic(list):
    n = len(list)
    for i in range(n):
        minimum = min(list[i:])
        min_index = list.index(minimum, i) # start searching from position i
        list[i], list[min_index] = list[min_index], list[i]
    return list

list_nums = [1,7,3,5,9,6]
selection_sort_pythonic(list_nums)
    
```

```python
def selection_sort_classic(list):
    for i in range(len(list)):
        min_idx = i
        for j in range(i + 1, len(list)):
            if list[i] < list[min_idx]:
                min_idx = i
            list[i], list[min_idx] = list[min_idx], list[i]
    return list

list_nums = [1,7,3,5,9,6]
selection_sort_classic(list_nums)
```

```python
import time
import sys
import random
list_of_nums = random.sample(range(1,100), 50)

def measure_time_complexity(func, list):
    start_time = time.time()
    func(list)
    end_time = time.time()
    return end_time - start_time

def measure_space_complexity(func, list):
    start_memory = sys.getsizeof(list)
    func(list)
    end_memory = sys.getsizeof(list)
    return end_memory - start_memory
```

```python
measure_time_complexity(selection_sort_classic, list_of_nums)
```

```python
import random
def merge(left, right):
    my_list = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            my_list.append(left[i])
            i +=1
        else:
            my_list.append(right[j])
            j += 1
        
    while i < len(left):
        my_list.append(left[i])
        i += 1
    
    while j < len(right):
        my_list.append(right[j])
        j += 1
    
    return my_list
        
def merge_sort(list):
    if len(list) <= 1:
        return list
    mid = len(list) // 2
    left_half = list[:mid]
    right_half = list[mid:]
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

list_of_nums = random.sample(range(50), 50)
y = merge_sort(list_of_nums)
    
```
